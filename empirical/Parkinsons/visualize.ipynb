```python
import json
```

```python
filename = "records.json"
with open(filename, "r") as f:
    records = json.load(f)
print(json.dumps(records, indent=2))
```

```python
not_none_count = {b: 0 for b in records["CoRT-SI"].keys()}
for key, value in records["CoRT-SI"].items():
    for v in value:
        if v is not None:
            not_none_count[key] += 1
```
