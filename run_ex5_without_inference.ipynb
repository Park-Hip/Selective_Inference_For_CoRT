```python
import ex5_cort_without_inference
import os
import datetime
import json
```

```python
n_source = 80
p = 200
K = 5
Ka = 2
h = 15
alpha = 0.05
T = 5
s_len = 10
s_vector = [0.5] * s_len
iteration = 1000

n_target_list = [50, 60, 70, 80]

for n_target in n_target_list:
    print(f"-- Processing n_target {n_target} --")
    fpr, tpr = ex5_cort_without_inference.cort_without_inference(iteration, n_target, n_source, p, K, Ka, h, alpha, T, s_len, s_vector)

    configs = {
        "n_source": n_source,
        "p":  p,
        "K": K,
        "Ka": Ka,
        "h": h,
        "alpha": alpha,
        "T":  T,
        "s_len": s_len,
        "s_vector": s_vector,
        "n_target": n_target,
        "iteration": iteration,
    }

    results = {
        "configs": configs,
        "time": datetime.datetime.now().ctime(),
        "fpr": fpr,
        "tpr": tpr
    }

    filename = 'records/without_inference_vis_records.json'
    if os.path.exists(filename):
        with open(filename, 'r') as f:
            try:
                records = json.load(f)
            except Exception as e:
                print(e)
                records = []
    else:
        records = []

    records.append(results)

    with open(filename, 'w') as f:
        json.dump(records, f, indent=4)
```
